# Plot first guess statistics for all glaciers and dynamic states

```python
glaciers = sorted(list(experiment_glaciers.keys()))
glacier_states = ['retreating', 'equilibrium', 'advancing']
fg_methods = ['fg_oggm', 'fg_glabtop']
lam = 'lam9'
obs = 'obs6'
main_results_dir = '/home/www/pschmitt/agile/final_runs_glacier_states/full_run/results_old/'
```

```python
for glacier in glaciers:
    print(glacier)
    for glacier_state in glacier_states:
        print('  ' + glacier_state)
        for fg_method in fg_methods:
            print('    ' + fg_method)
            mad_today_volume_ref, mad_bed_h_ref, mad_past_volume_ref = get_ref_stats(
                main_results_dir, glacier, glacier_state,
                lam, obs, fg_method, reg='reg0')

            print(f'      MAD_BED: {mad_bed_h_ref:.1f} m')
            print(f'      MAD_V_1980: {mad_past_volume_ref/1e6:.2f} 1e6 m3')
            print(f'      MAD_V_2020: {mad_today_volume_ref/1e6:.2f} 1e6 m3')
```
